```python
import os, sys, time, glob, warnings
from IPython.display import clear_output             # use inside loop with clear_output(wait = True) followed by print(i)
warnings.filterwarnings('ignore')
this_dir = os.getcwd()
data_dir = this_dir + '/../../data'                  # large datasets reside outside the repository

from matplotlib import pyplot as plt
from matplotlib import colors as mplcolors
import numpy as np, pandas as pd, xarray as xr
from numpy import datetime64 as dt64, timedelta64 as td64

# convenience functions abbreviating 'datetime64' and so on
def doy(theDatetime): return 1 + int((theDatetime - dt64(str(theDatetime)[0:4] + '-01-01')) / td64(1, 'D'))
def dt64_from_doy(year, doy): return dt64(str(year) + '-01-01') + td64(doy-1, 'D')
def day_of_month_to_string(d): return str(d) if d > 9 else '0' + str(d)

print('\nJupyter Notebook running Python {}'.format(sys.version_info[0]))
```

```python
# Function is duplicated in notebook Ocean 01 G
# If modified here: Please re-copy it there.

def ReadProfiles(fnm):
    """
    Profiles are saved by site and year as 12-tuples. Here we read only
    the datetimes (not the indices) so there are only six values. These
    are converted to Timestamps. They correspond to ascend start/end, 
    descend start/end and rest start/end.
    """
    df = pd.read_csv(fnm, usecols=["1", "3", "5", "7", "9", "11"])
    df.columns=['ascent_start', 'ascent_end', 'descent_start', 'descent_end', 'rest_start', 'rest_end']
    df['ascent_start'] = pd.to_datetime(df['ascent_start'])
    df['ascent_end'] = pd.to_datetime(df['ascent_end'])
    df['descent_start'] = pd.to_datetime(df['descent_start'])
    df['descent_end'] = pd.to_datetime(df['descent_end'])
    df['rest_start'] = pd.to_datetime(df['rest_start'])
    df['rest_end'] = pd.to_datetime(df['rest_end'])
    return df

df_osb2021 = ReadProfiles(os.getcwd()+"/profiles/osb2021.csv")
df_axb2021 = ReadProfiles(os.getcwd()+"/profiles/axb2021.csv")

```

```python
# generate four dataframes of profiles: (osb, axb) x (Jan 2021, Jul 2021)
tRangeJan1, tRangeFeb1 = dt64('2021-01-01'), dt64('2021-02-01')
tRangeJul1, tRangeAug1 = dt64('2021-07-01'), dt64('2021-08-01')

df_osb_jan = df_osb2021[(df_osb2021['ascent_start'] > tRangeJan1) & (df_osb2021['ascent_start'] < tRangeFeb1)]
df_osb_jul = df_osb2021[(df_osb2021['ascent_start'] > tRangeJul1) & (df_osb2021['ascent_start'] < tRangeAug1)]
df_axb_jan = df_axb2021[(df_axb2021['ascent_start'] > tRangeJan1) & (df_axb2021['ascent_start'] < tRangeFeb1)]
df_axb_jul = df_axb2021[(df_axb2021['ascent_start'] > tRangeJul1) & (df_axb2021['ascent_start'] < tRangeAug1)]

# These four dataframes' data consists of timestamps (but not dt64)
# Here are some notes on diagnostic printouts:
#     Bit unwieldy, but can: print("\n\nosb january:\n\n", df_osb_jan)
#     Duration of first ascent: print(df_osb_jan['ascent_end'][0] - df_osb_jan['ascent_start'][0])
#     This works as dt64 to print an example time-of-ascent
#         duration = dt64(df_osb_jan['ascent_end'][0]) - dt64(df_osb_jan['ascent_start'][0])
#         print(duration.astype('timedelta64[m]'))
#         print(type(gap_in_time))
# 
# From 'print(df_osb_jan.shape[0])'' etcetera:
# 
# osb, jan has 256 profiles of 279 possible
# osb, jul     271
# axb, jan     254
# axb, jul     278
```

```python
# Characterize ascent times for Oregon Slope Base (osb), January 2021.
# One missing thing: Is a fast ascent stopping below the nominal depth?

# Calculate ascent durations for osb/january as pandas Dataseries
durations_osb_jan_series = df_osb_jan['ascent_end'] - df_osb_jan['ascent_start']
n_ascents = len(durations_osb_jan_series)
durations_osb_jan_mean = td64(durations_osb_jan_series.mean())

print('\nthere are', n_ascents, 'ascents for osb in jan 2021, mean duration', \
      durations_osb_jan_mean/td64(1, "m"), '\n')

# Prepare to scan all 256 profiles for anomalous time durations: outside (mean +- threshold) minutes
mdenom = td64(1, "m")
mean_ascent_time = durations_osb_jan_mean/mdenom            # converts to float (in minutes)
the_threshold = 5.

# further the diagnostics
nHi, nLo, minVal, maxVal = 0, 0, 1000000000000000000, 0
for i in range(n_ascents):
    ascent_duration = durations_osb_jan_series[i]/mdenom
    if ascent_duration > mean_ascent_time + the_threshold:
        nHi += 1
        if nHi == 1: print('First Hi: ', i, durations_osb_jan_series[i]/mdenom)
    if ascent_duration < mean_ascent_time - the_threshold:
        nLo += 1
        if nLo == 1: print('First Lo: ', i, durations_osb_jan_series[i]/mdenom)
    if ascent_duration > maxVal: maxVal = ascent_duration
    if ascent_duration < minVal: minVal = ascent_duration
        
print("\nof", n_ascents, "ascents there were", nLo, "(arbitrary threshold) low values,", nHi, "high\n")
print("fastest ascent was", minVal, "minutes\n")
print("slowest ascent was", maxVal, "minutes\n")  
print("future: It would be interesting to know if fast ascents top out deeper.\n")
```

## What happens next

For each of the two sites and the two time ranges:

- Read in a full dataset for, say, temperature as an XArray Dataset
- Per sensor type: Clean up the data variables (renames, drops)
- Accumulate a list of sub-Datasets that fit into the ascent time ranges
- Concatenate these into a single Dataset
- Write that as a NetCDF file inside this repository

```python
# df_osb_jan, df_axb_jul
```

```python
# just use df_osb_jan
sitelist = ['osb', 'axb']
monthlist = ['jul21']
sensorlist = ["temperature"]
df_timestamps = [df_osb_jul, df_axb_jul]

for site_index in range(len(sitelist)):
    for month_index in range(len(monthlist)):
        for sensor_index in range(len(sensorlist)):
            
            
            inputfile = os.getcwd() + '/../data/data_explorer_1Min/' + sitelist[site_index] + '/profiler/' 
            inputfile = inputfile + sitelist[site_index] + '_profiler_' + sensorlist[sensor_index] + "_1Min.nc"

            print("\nsite", site_index, "month", month_index, "sensor", sensor_index, 'input', inputfile, "\n")
            
            ds = xr.open_dataset(inputfile)
            
            if sensorlist[sensor_index] == 'temperature':
                ds = ds.rename({'sea_water_temperature_profiler_depth_enabled':'temperature'})
                ds = ds.drop('sea_water_temperature_profiler_depth_enabled_qc_agg')
     
            time_df = df_timestamps[month_index * 2 + site_index]
        
            # print(time_df)

            n_ascents = len(time_df)
            
            for i in range(n_ascents):
                print('of', n_ascents, 'at', i, 'ascent_start =', time_df['ascent_start'][i])
                
            ds_concat = xr.concat([ds.sel(time=slice(                                     \
                                                      dt64(time_df['ascent_start'][i]),   \
                                                      dt64(time_df['ascent_end'][i])      \
                                                    )                                     \
                                         ) for i in range(n_ascents)                      \
                                  ], dim='time')
            
            outputfile = os.getcwd() + '/data/rca/redux/' + sitelist[site_index] + '/' + monthlist[month_index] + '/'
            outputfile = outputfile + sensorlist[sensor_index] + '.nc'
            ds_concat.to_netcdf(outputfile)
```

```python
df_timestamps = [[df_osb_jan, df_osb_jul],[df_axb_jan, df_axb_jul]]
sitelist = ['osb', 'axb']
monthlist = ['jan21', 'jul21']
# sensorlist = ["backscatter", "cdom", "chlora", "density", "doxygen", "nitrate", "par", "pco2", \
#               "ph", "pressure", "salinity", "spkir", "temperature", "veleast", "velnorth", "velup"]
sensorlist = ["temperature"]   # , "backscatter"]

for site_index in range(len(sitelist)):
    for month_index in range(len(monthlist)):
        for sensor_index in range(len(sensorlist)):
            
            inputfile = os.getcwd() + '/../data/data_explorer_1Min/' + sitelist[site_index] + '/profiler/' 
            inputfile = inputfile + sitelist[site_index] + '_profiler_' + sensorlist[sensor_index] + "_1Min.nc"
            
            ds = xr.open_dataset(inputfile)
            
            # customized by sensor type: rename some data variables, drop extraneous ones
            if sensorlist[sensor_index] == 'backscatter':
                ds = ds.rename({'flubsct_profiler_depth_enabled':'backscatter'})
            if sensorlist[sensor_index] == 'temperature':
                ds = ds.rename({'sea_water_temperature_profiler_depth_enabled':'temperature'})
                ds = ds.drop('sea_water_temperature_profiler_depth_enabled_qc_agg')
           
            # df_time points to the correct data frame with ascent start/stop times
            df_time = df_timestamps[site_index][month_index].copy()
            n_ascents = len(df_time)
            print(df_time, '\n\n\n', n_ascents)
            
            # Uses a list comprehension to select out many Datasets corresponding to the many ascents.
            # These are concatenated together into a single Dataset
            ds_concat = xr.concat([ds.sel(time=slice(                                     \
                                                      dt64(df_time['ascent_start'][i]),   \
                                                      dt64(df_time['ascent_end'][i])      \
                                                    )                                     \
                                         ) for i in range(n_ascents)                      \
                                  ], dim='time')
            
            # The output file name is manufactured from the lists and indices at the top
            outputfile = os.getcwd() + '/data/rca/redux/' + sitelist[site_index] + '/' + monthlist[month_index] + '/'
            outputfile = outputfile + sensorlist[sensor_index] + '.nc'
            
            # These files are intended to come in at < 1/2 MB each (for the CDT and fluorometer)
            #   Other sensor types will vary 
            # There are two sites and two months; so that is a factor of 4.  
            ds_concat.to_netcdf(outputfile)
```
